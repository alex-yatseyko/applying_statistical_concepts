## Ukraine Cities Population

```python
import pandas as pd

df = pd.read_csv('Ukraine_Major_City_Population.csv')

df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))

cities = df['City'].unique()

#df['Population'] = pd.to_numeric(df['Population'].str.replace(',', '').astype(int))

for city in cities:
    city_data = df[df['City'] == city]
    plt.plot(city_data['Year'], city_data['Population'], marker='o', linestyle='-', label=city)

max_population = df['Population'].max()


plt.xlabel('Year')
plt.ylabel('Population')
plt.yticks(range(0, max_population, 1000000), rotation=45)

plt.title('Population Trends of Cities')
plt.legend()

plt.grid(True)

plt.tight_layout()

plt.show()

```

```python
# The graph above is not really representative. Too many cities and lines.
# Let's focus on Kyiv in the beginning and compare it to Kharkiv and Odesa for example

kyiv = df[df['City'] == 'Kyiv']
kharkiv = df[df['City'] == 'Kharkiv']
odesa = df[df['City'] == 'Odesa']

plt.plot(kyiv['Year'], kyiv['Population'], marker='o', linestyle='-', label='Kyiv')
plt.plot(kharkiv['Year'], kharkiv['Population'], marker='o', linestyle='-', label='Kharkiv')
plt.plot(odesa['Year'], odesa['Population'], marker='o', linestyle='-', label='Odesa')

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Trends of Cities')
plt.legend()

plt.grid(True)

plt.tight_layout()

plt.show()
```
